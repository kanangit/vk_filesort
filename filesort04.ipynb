```python
new_excelFilePath = r"i:/OneDrive/temp/forVK/fileSortingPrj/data/Photos_Consolidated_processed.xlsx"
not_exist_excelFilePath = r"i:/OneDrive/temp/forVK/fileSortingPrj/data/Photos_Consolidated_nonExistent.xlsx"
```

```python
df["Photo file name (CT_Number):"] = df["Photo file name (CT_Number):"].astype(str)
```

```python
df["file_exist"] = False
for i, row in df.iterrows():
    base = row["Photo file name (CT_Number):"]
    for ext in (".png", ".jpeg", ".jpg", ".PNG", ".JPEG", ".JPG"):
        if os.path.exists(os.path.join(dir_phot_path, base + ext)):
            df.at[i, "file_exist"] = True
            break
```

```python
df_not_exist = df[df["file_exist"] == False]
```

```python
df["filename_new"] = df["Photo file name (CT_Number):"].str.replace(r"(\d+)", r"0000\1", regex=True)
```

```python
df["filename_new"] = df["filename_new"].str.replace(r"0*(\d{4,})", r"\1", regex=True)
```

```python
df["sel_flag"] = "undefined"
```

```python
df["sel_flag"] = "undefined"

# work only with rows where file_exist == True
exist_idx = df.index[df["file_exist"] == True]

# --- 1) COMMON ---
# assumes 500 <= number of existing files; adjust if needed
common_idx = df.loc[exist_idx].sample(n=500, random_state=42).index
df.loc[common_idx, "sel_flag"] = "common"

# --- 2) Remaining after 'common' among existing files ---
remaining_idx = exist_idx.difference(common_idx)

# --- 3) Eligible for A: remaining & Outlet != "Citizen Times" ---
eligible_v_idx = df.loc[remaining_idx][df.loc[remaining_idx, "Outlet"] == "Citizen Times"].index

# Target size for V: only "Citizen Times" files
target_v = len(remaining_idx) // 2

v_idx = eligible_v_idx[:target_v]

# --- 4) A = all other remaining existing files ---
a_idx = remaining_idx.difference(v_idx)

df.loc[a_idx, "sel_flag"] = "A"
df.loc[v_idx, "sel_flag"] = "V"

```

```python
len(df[df["sel_flag"] == "V"])
```

```python
for i, row in df.iterrows():
    base_old = row["Photo file name (CT_Number):"]
    base_new = row["filename_new"]
    for ext in (".png", ".jpeg", ".jpg", ".PNG", ".JPEG", ".JPG"):
        old_path = os.path.join(dir_phot_path, base_old + ext)
        if os.path.exists(old_path):
            new_path = os.path.join(dir_phot_path, base_new + ext)
            os.rename(old_path, new_path)
            break
        else:
            print(f"Missing: {base_old}")
            
```

```python
for i, row in df.iterrows():
    fname = row["filename_new"]

    # try both extensions
    for ext in (".png", ".jpeg", ".jpg", ".PNG", ".JPEG", ".JPG"):
        src = os.path.join(dir_phot_path, fname + ext)
        if os.path.exists(src):
            flag = row["sel_flag"]
            dest_dir = os.path.join(dir_phot_path, flag)

            # create subfolder if it doesnt exist
            os.makedirs(dest_dir, exist_ok=True)

            dest = os.path.join(dest_dir, fname + ext)
            shutil.move(src, dest)
            break
    else:
        print(f"Missing: {fname}")
```
